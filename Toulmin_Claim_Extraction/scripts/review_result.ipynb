```python
OUTPUT_FILE = r'../results/sample_500_meta_result_format_result.jsonl'
path = Path(OUTPUT_FILE)
records = []
with open(path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        try:
            records.append(json.loads(line))
        except json.JSONDecodeError:
            print(line)

df = pd.DataFrame(records)

responses = df[1]
metadata = df[2]

records = []
for i in range(len(df)):
    content = responses[i]['choices'][0]['message']['content']
    
    try:
        dct = json.loads(content)
    except:
        continue
    claim = dct.get("Claim", '')
    if claim is None:
        claim = dct.get('claim', '')
    ground = dct.get('Ground', '')
    warrant = dct.get('Warrant', '')

    md = metadata[i]
    input = md['input']
    input_prompt = md['input_prompt']

    mdmd = md.get('metadata', '')
    mdmd = ast.literal_eval(mdmd)
    llm_theme_human_truth = mdmd.get('llm_theme_human_truth', '')
    llm_theme_llm_truth = mdmd.get('llm_theme_llm_truth', '')
    llm_theme = mdmd.get('llm_theme_theme', '')
    
    row = {'input': input, 'input_prompt': input_prompt, 'response': content, 'claim': claim, 'ground': ground, 'warrant': warrant, 
           'llm_theme_human_truth': llm_theme_human_truth, 
           'llm_theme_llm_truth': llm_theme_llm_truth,
           'llm_theme': llm_theme,
           'metadata': md}
    records.append(row)

out_df = pd.DataFrame(records)
save_dir = path.parent
out_df.to_csv(f"{save_dir}/sample_500_meta_result_format_result.csv")
```

```python
SYSTEM_PROMPT = """You are an information extraction system for Toulmin argument components.

Definitions:
- Claim: a disputable position, judgment, or persuasive statement.
- Ground: evidence or factual support explicitly stated in the text.
- Warrant: a reasoning statement that explicitly explains why the ground supports the claim.

Important rules (read carefully):
- Extract ONLY what is explicitly stated.
- Do NOT infer, assume, or add common-sense reasoning.
- If a component is not explicitly written in the text, output "Not present".
- Warrants are RARE. Most texts do NOT contain an explicit warrant.
- Do NOT create a warrant unless the reasoning is clearly and explicitly stated.
- If the text is purely factual, descriptive, or promotional without a disputable position, the claim is "Not present".
- Return STRICT JSON only:
{"claim": "...", "ground": "...", "warrant": "..."}"""

FEWSHOT_BLOCK = """
Example 1
Text:
Upgrade your Teaching Toolbox! Uncover free interactive tools with standout resources for teaching fiscal policy. Transform your lessons into engaging, unforgettable experiences.

Output:
{"claim":"Free interactive tools can transform fiscal policy lessons into engaging and effective learning experiences.","ground":"Not present","warrant":"Not present"}

Example 2
Text:
Accelerating the future of energy, together

Output:
{"claim":"Not present","ground":"Not present","warrant":"Not present"}

Example 3
Text:
Our spray foam insulation helps reduce energy waste and supports a more sustainable future.

Output:
{"claim":"Using this insulation supports a more sustainable future.","ground":"Our spray foam insulation helps reduce energy waste.","warrant":"Not present"}

Example 4 (explicit warrant case)
Text:
Our insulation reduces energy waste. Because reducing energy waste lowers carbon emissions, it helps protect the environment.

Output:
{"claim":"Using this insulation helps protect the environment.","ground":"Our insulation reduces energy waste.","warrant":"Reducing energy waste lowers carbon emissions."}
""".strip()
```

```python
def build_user_prompt(ad_text: str) -> str:
    return f"""{FEWSHOT_BLOCK}

Now analyze the following text.

Text:
{ad_text}

Output:
""".strip()
```

```python
def extract_key(toulmin_json, key = "claim"):
    if pd.isna(toulmin_json):
        return "Not present"
    s = str(toulmin_json).strip()
    if not s:
        return "Not present"
    try:
        obj = json.loads(s)
        if isinstance(obj, dict):
            claim = obj.get(key, "Not present")
            claim = "Not present" if claim is None else str(claim).strip()
            return claim if claim else "Not present"
    except Exception:
        pass
    return "Not present"
```

```python
claims = []
grounds = []
warrants = []

no_ground = 0
with_claims = 0
no_ground_with_claims = 0
with_warrant = 0
for pred in df['pred_toulmin_json']:
    claim = extract_key(pred)
    ground = extract_key(pred, key = "ground")
    warrant = extract_key(pred, key = "warrant")

    claims.append(claim)
    grounds.append(ground)
    warrants.append(warrant)

    if claim != "Not present":
        with_claims += 1
    
    if ground == "Not present":
        no_ground += 1
    
    if claim != "Not present" and ground == "Not present":
        no_ground_with_claims += 1
    
    if warrant != "Not present":
        with_warrant += 1


print(with_claims / len(df))
print(with_warrant / len(df))
print(no_ground / len(df))
print(no_ground_with_claims / len(df))
```
